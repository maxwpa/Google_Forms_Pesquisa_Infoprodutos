```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Dados_Tratados.csv')
```

```python
df.describe().round(2)
```

```python
qualy = df.select_dtypes(exclude=[np.number])
for coluna in qualy.columns:
    resul = pd.DataFrame(qualy[coluna].value_counts(normalize=True).map(lambda x: '{:.2%}'.format(x)))
    newname = {'proportion': 'Proporção'}
    resul.rename(columns=newname, inplace=True)
    display(resul)
    print('\n' + '-'*30 + '\n')
```
