```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Dados_Tratados.csv')
```

```python
qualy = df.select_dtypes(exclude=[np.number])
quanty = df.select_dtypes(exclude=[object])
```

```python
for i in range(len(qualy.columns)):
    for j in range(i+1, len(qualy.columns)):
        tab_qualy = pd.crosstab(index=qualy[qualy.columns[i]], columns=qualy[qualy.columns[j]], normalize='index').applymap('{:.2%}'.format).T
        display(tab_qualy)
        print('\n' + '-'*30 + '\n')
```

```python
for col_qualy in qualy.columns:
    for col_quanty in quanty.columns:
        if col_qualy != col_quanty:
            result = df.groupby(by=col_qualy)[col_quanty].agg(
                Média='mean', Mediana='median', DesvPad='std').applymap('{:.2}'.format)
            result = result.rename_axis(col_quanty, axis=1)
            display(result)
            print('\n' + '-'*30 + '\n')
```
